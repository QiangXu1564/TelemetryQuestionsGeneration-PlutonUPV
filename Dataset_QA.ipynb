```python
from datasets import load_dataset
import json

dataset = load_dataset("plutonupv/Estigia_catalan", split="train")

dataset_formateado = []

print("Procesando y extrayendo texto en inglés...")

for item in dataset:

    conv = item.get("conversation", [])
    
    if conv and len(conv) > 0:
        pregunta_en = conv[0].get("content", "")

        nuevo_objeto = {
            "Question": pregunta_en,
            "Label": "GENERAL_CHAT",
            "Language": "en"
        }
        dataset_formateado.append(nuevo_objeto)

with open("dataset_en_estigia.json", "w", encoding="utf-8") as f:
    json.dump(dataset_formateado, f, ensure_ascii=False, indent=4)

print(f"Éxito: Se han guardado {len(dataset_formateado)} registros en inglés.")
```
